```python
# Steps that I took to clean the data and preprocess the data
# Dropping unnecessary columns
# Convert TotalCharges to numeric
# Handle missing values
# I used label encoding for Binary Categorical Columns
# I used one-hot encoding for multi-class categorical columns
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'
data = pd.read_csv(file_path)
```

```python
data.head()
```

```python
data.info()
```

```python
# Dropping the unnecessary columns
data_cleaned = data.drop(columns=["customerID"])
```

```python
# Convert 'TotalCharges' to numeric, coerce errors to NaN
data_cleaned["TotalCharges"] = pd.to_numeric(data_cleaned["TotalCharges"], errors="coerce")
```

```python
# Handle missing values in 'TotalCharges' (replace with the median)
data_cleaned["TotalCharges"].fillna(data_cleaned["TotalCharges"].median(), inplace=True)
```

```python
# Label encode binary categorical columns
binary_columns = [
    "gender", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"
]
label_encoders = {}
for col in binary_columns:
    le = LabelEncoder()
    data_cleaned[col] = le.fit_transform(data_cleaned[col])
    label_encoders[col] = le
```

```python
# One-hot encode multi-class categorical columns
multi_class_columns = [
    "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup",
    "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies",
    "Contract", "PaymentMethod"
]
data_encoded = pd.get_dummies(data_cleaned, columns=multi_class_columns, drop_first=True)
```

#### Note that one-hot encoding columns are treated as boolean, meaning they are effectively stored as 1s and 0s in the underlying data type

```python
# Export the cleaned and encoded dataset to a CSV file
export_path = 'cleaned_telco_churn_data.csv'
data_encoded.to_csv(export_path, index=False)
```
